```python
def extract_audio_tracks(mp4_file, output_dir):
    audio_tracks = []
    for i in range(5):  # Assuming there are 5 audio tracks (0 to 4)
        output_file = os.path.join(output_dir, f'{os.path.basename(mp4_file)}_track{i}.mp3')
        subprocess.run(['ffmpeg', '-i', mp4_file, '-map', f'0:a:{i}', output_file])
        audio_tracks.append(output_file)
    return audio_tracks
```

```python
def process_audio_track(audio_file, sr=22050, duration=1):
    y, _ = librosa.load(audio_file, sr=sr)
    total_length = len(y)
    block_length = sr * duration
    blocks = []

    for i in range(0, total_length, block_length):
        block = y[i:i + block_length]
        if len(block) < block_length:
            block = np.pad(block, (0, block_length - len(block)), mode='constant')
        blocks.append(block)

    return blocks
```

```python
def perform_stft(y, n_fft=500, hop_length=125):   #n_fft=2048 hop_length=125 too big for computation
    return librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
```

```python

def build_cnn_model(input_shape):
    with strategy.scope():
        inputs = Input(shape=input_shape)

        x = Conv2D(32, kernel_size=(3, 3), padding='same')(inputs)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(64, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(128, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(256, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(512, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(256, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(128, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(64, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(32, kernel_size=(3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        # Output layer with linear activation
        outputs = Conv2D(8, kernel_size=(3, 3), padding='same', activation='linear')(x)

        model = Model(inputs, outputs)
    return model
```

```python
def train_model(model, X_train, y_train, epochs=10, batch_size=16, validation_split=0.2):
    with strategy.scope():
        model.compile(optimizer='adam', loss='mean_squared_error')
        model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=validation_split)
```

```python
def evaluate_model(model, X_test, y_test):
    loss = model.evaluate(X_test, y_test, verbose=0)
    print(f'Test loss: {loss:.4f}')
```

```python
def memory_usage(message: str = ''):
    process = psutil.Process(os.getpid())
    mem_info = process.memory_info()
    print(f'{message} - RSS: {mem_info.rss / 1024 ** 2:.2f} MB')
```

```python
import tensorflow.keras.backend as K


def clear_memory():
    K.clear_session()
    tf.compat.v1.reset_default_graph()
    gc.collect()
```

```python
def prepare_complex_data(data):
    data_real = np.real(data)
    data_imag = np.imag(data)
    data_combined = np.concatenate((data_real, data_imag), axis=-1)
    return data_combined
```

```python
from memory_profiler import profile

@profile
def train_model_with_profile():
    import warnings
    warnings.filterwarnings('ignore')
    import tensorflow as tf
    tf.get_logger().setLevel('ERROR')

    i=66
    for mp4_file in mp4_files[66:]:
        mp4_path = os.path.join(input_dir, mp4_file)
        audio_tracks = extract_audio_tracks(mp4_path, output_dir)

        data = []


        # Initialize a dictionary to store blocks for each track
        blocks_dict = {'file': mp4_file}

        for track_idx, audio_file in enumerate(audio_tracks):
            blocks = process_audio_track(audio_file)
            for block_idx, block in enumerate(blocks):
                stft_matrix = perform_stft(block)

                # Create keys for each block index if not exists
                if f'block_{block_idx}' not in blocks_dict:
                    blocks_dict[f'block_{block_idx}'] = {}

                blocks_dict[f'block_{block_idx}'][f'track_{track_idx}'] = stft_matrix

        # Append the dictionary to data
        data.append(blocks_dict)

        del blocks, blocks_dict,audio_tracks,stft_matrix
        gc.collect()

        for root, dirs, files in os.walk(output_dir, topdown=False):
            for name in files:
                os.remove(os.path.join(root, name))
            for name in dirs:
                os.rmdir(os.path.join(root, name))


        # Create a list of dictionaries with consistent keys for DataFrame
        df_data = []
        for entry in data:
            file = entry['file']
            for block_key in entry:
                if block_key == 'file':
                    continue
                block_idx = int(block_key.split('_')[1])
                row = {'file': file, 'block': block_idx}
                row.update(entry[block_key])
                df_data.append(row)

        # Create a dataframe
        df_batch = pd.DataFrame(df_data)

        del df_data
        gc.collect()

        X = np.stack(df_batch['track_0'].values)  
        X = np.expand_dims(X, axis=-1)  

        y = np.stack([np.stack(df_batch['track_1'].values),
                      np.stack(df_batch['track_2'].values),
                      np.stack(df_batch['track_3'].values),
                      np.stack(df_batch['track_4'].values)], axis=-1)  

        del df_batch
        gc.collect()


        X_prepared = prepare_complex_data(X)  
        y_prepared = prepare_complex_data(y)  

        del X,y
        gc.collect()

        # Split data into train and test sets
        X_train, X_test, y_train, y_test = train_test_split(X_prepared, y_prepared, test_size=0.2, random_state=42)
        X_train = X_train.astype(np.float32)
        y_train = y_train.astype(np.float32)
        X_test = X_test.astype(np.float32)
        y_test = y_test.astype(np.float32)

        del X_prepared,y_prepared
        gc.collect()

        memory_usage('after training data')
        # Clear memory: delete variables and perform garbage collection

    #     print(X_train)
    #     print(X_train.shape)

        model_file = f'model_batch_{i-1}.h5'
        if os.path.exists(model_file):
            # Load the existing model
            with strategy.scope():
                model = keras.models.load_model(model_file)
                print(f"Continuing training from {model_file}")
        else:
            # Create a new model if no saved model found
            input_shape = (251,177, 2)
            with strategy.scope():
                model = build_cnn_model(input_shape)
                model.compile(optimizer='adam', loss='mean_squared_error')
                print("Training a new model...")



    #     Train the model
        with strategy.scope():
            train_model(model, X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

        evaluate_model(model, X_test, y_test)

    #     Save the model (adjust filename as needed)
        model.save(f'model_batch_{i}.h5')

        # Clear memory
        del X_train, y_train, X_test, y_test
        gc.collect()

        i+=1
        memory_usage('after model')
        del model
        gc.collect()

        tf.keras.backend.clear_session()
        clear_memory()

```

```python
import scipy.signal

def apply_wiener_filter(stft_matrix):
    # Apply Wiener filter to each 2D slice of the STFT matrix
    filtered_stft = np.zeros_like(stft_matrix)
    for i in range(stft_matrix.shape[-1]):  # Iterate over the last dimension (4 in your case)
        filtered_stft[..., i] = scipy.signal.wiener(stft_matrix[..., i], (5, 5))
    return filtered_stft

# Apply the Wiener filter to each predicted STFT matrix
for i in range(predicted_array.shape[-1]):
    predicted_array[..., i] = apply_wiener_filter(predicted_array[..., i])


```

```python
import numpy as np
import librosa
import soundfile as sf
from pydub import AudioSegment

# Example complex array (replace this with your actual complex array)
# predicted_array = np.random.rand(75, 251, 177, 4) + 1j * np.random.rand(75, 251, 177, 4)

# Original frame rate of the audio
original_sr = 48000

# Define the function to convert complex array back to time-domain signal using ISTFT
def istft(stft_matrix):
    return librosa.istft(stft_matrix,n_fft=500,hop_length=125)

# Convert the complex array back to a time-domain signal using ISTFT
signals = [[] for _ in range(4)]  # Create a list of lists to store signals for each track
for i in range(187):
    for j in range(4):
        track_complex = predicted_array[i, ..., j]
        # Perform inverse STFT for each track
        signal = istft(track_complex)
        # Resample to the original frame rate
        signal_resampled = librosa.resample(signal, orig_sr=22050, target_sr=original_sr)
        signals[j].append(signal_resampled.astype(np.float32))  # Ensure the data type is float32

# Concatenate the signals for each track into one long track
long_tracks = []
for i in range(4):
    long_track = np.concatenate(signals[i], axis=0)
    long_tracks.append(long_track)

# Save each long track as a separate MP3 file
for i, long_track in enumerate(long_tracks):
    wav_file = f'track2_wiener_{i}.wav'
    sf.write(wav_file, long_track, original_sr, subtype='PCM_16')


print("Audio tracks have been saved as separate MP3 files.")
```
